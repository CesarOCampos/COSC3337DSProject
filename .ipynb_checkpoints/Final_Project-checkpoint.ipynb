```python
print("The goal is to predict if a song will be a hit. 0 = No, 1 = Yes. In the dataset it's labeled: \"Target\".")
```

```python
print("How many hits vs non-hits?")
```

```python
s.countplot(x = df['target'], data= df)
plt.show()
print("It's about 50% hits and 50% flops")
```

```python
print("Now we create the Logistic Regression model to predict hit/flop.")
```

```python
from sklearn.model_selection import train_test_split as tts

x = df[['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness','liveness','valence','tempo','duration_ms','time_signature','chorus_hit','sections']]
y = df['target']

xtrain, xtest, ytrain, ytest = tts(x, y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression

logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(xtrain, ytrain)
```

```python
predictions = logmodel.predict(xtest)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print(classification_report(ytest, predictions))
print(confusion_matrix(ytest, predictions))
```
